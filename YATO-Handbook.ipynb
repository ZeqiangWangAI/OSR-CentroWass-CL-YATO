```python
import matplotlib.pyplot as plt
import pandas as pd
import math

# 数据处理函数
def process_data(text):
    lines = text.strip().split('\n')
    data = [line.split(', F1 Score: ') for line in lines]
    df = pd.DataFrame(data, columns=['Class', 'F1 Score'])
    df['F1 Score'] = pd.to_numeric(df['F1 Score'], errors='coerce')
    return df.set_index('Class')



# 合并数据
merged_data = pd.concat([data1, data2], axis=1, keys=['Group 1', 'Group 2'])
merged_data = merged_data.sort_values(by=('Group 1', 'F1 Score'), ascending=False)

# 定义每个图表显示的类别数量
classes_per_chart = 20

# 计算需要的图表数量
num_charts = math.ceil(len(merged_data) / classes_per_chart)

# 创建并保存多个图表
for i in range(num_charts):
    start_idx = i * classes_per_chart
    end_idx = (i + 1) * classes_per_chart
    
    plt.figure(figsize=(15, 10))
    
    data_subset = merged_data.iloc[start_idx:end_idx]
    
    data_subset['Group 1'].plot(marker='o', linestyle='-', linewidth=2, markersize=5)
    data_subset['Group 2'].plot(marker='s', linestyle='--', linewidth=2, markersize=5)
    
    plt.title(f'F1 Score Comparison (Chart {i+1}/{num_charts})', fontsize=16)
    plt.xlabel('Classes', fontsize=12)
    plt.ylabel('F1 Score', fontsize=12)
    plt.legend(['Group 1', 'Group 2'], fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.ylim(0, 1)
    
    plt.xticks(range(len(data_subset.index)), data_subset.index, rotation=45, ha='right', fontsize=8)
    
    plt.tight_layout()
    
    plt.savefig(f'f1_score_comparison_chart_{i+1}.png', dpi=300, bbox_inches='tight')
    
    plt.close()

print(f"已生成 {num_charts} 张图表。")
```
